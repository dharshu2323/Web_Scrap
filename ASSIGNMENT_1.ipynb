### I HAVE CHOOSEN CARS24 WEBSITE FOR SCRAPING


```python
url="https://www.cars24.com/buy-used-hyundai-cars-chennai/"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BS(response.text)
```

```python
soup
```

```python
name=soup.select("._3FpCg")
```

```python
name
```

```python
Car_Model_Name=[]
for i in name:
    Car_Model_Name.append(i.text)
```

```python
Car_Model_Name
```

```python
fuel=soup.select(".bVR0c li~ li+ li")
```

```python
fuel
```

```python
Fuel_Type=[]
for j in fuel:
    Fuel_Type.append(j.text)
```

```python
Fuel_Type
```

```python
price=soup.select("._7udZZ span")
```

```python
price
```

```python
Fixed_Price=[]
for k in price:
    Fixed_Price.append(k.text)
```

```python
Fixed=Fixed_Price[0:20]
```

```python
Fixed
```

```python
emi=soup.select("strong")
```

```python
emi
```

```python
EMI=[]
for l in emi:
    EMI.append(l.text)
```

```python
EMI
```

```python
emi=EMI[0:20]
```

```python
emi
```

```python
owner=soup.select(".bVR0c li:nth-child(2)")
```

```python
owner
```

```python
Owner=[]
for y in owner:
    Owner.append(y.text)
```

```python
Owner
```

```python
kilometer=soup.select(".bVR0c li:nth-child(1)")
```

```python
kilometer
```

```python
Kilometer_Driven=[]
for v in kilometer:
    Kilometer_Driven.append(v.text)
```

```python
Kilometer_Driven
```

```python
kilo=Kilometer_Driven[0:20]
```

```python
kilo
```

```python
transmission=soup.select(".cvakB span")
```

```python
transmission
```

```python
Transmission_Type=[]
for u in transmission:
    Transmission_Type.append(u.text)
```

```python
Transmission_Type
```

```python
import pandas as pd
```

```python
car_data=pd.DataFrame(Car_Model_Name,columns=["CAR_MODEL_NAME"])
```

```python
car_data["FIXED_PRICE"]=Fixed
```
